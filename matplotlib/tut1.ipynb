### Pyplot functions
1. plt.show()
2. plt grid()
3. plt.legend()
4. plt.title()
5. plt.plot()  for constructing line graphs
6. plt.xlim(), plt.ylim()

plt.plot(x, y, color="color name", linestyle="line-style", linewidht=val, marker="marker", label="label name")

```python
months = [1,2,3,4]
sales = [1000, 1500, 1200, 1800]

plt.plot(months, sales, color = "blue", linestyle="--", linewidth=2, marker="o", label="2025 sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("2024 sales data")
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.legend(loc='upper right', fontsize=12)  #upper left(by default)
plt.xlim(1, 4)
plt.ylim(0, 2000)
plt.xticks([1,2,3,4], ['m1', 'm2', 'm3', 'm4'])
plt.show()
```

### Barplot
plt.bar(x, height, color='color', width=value, label='label name')

```python
prod = ['A', 'B', 'C', 'D']
sale = [1000, 1500, 800, 1200]

plt.bar(prod, sale, color="orange", label="Sales 2025")
plt.xlabel("product")
plt.ylabel("Sales")
plt.legend()
plt.title("Product sales comparison")
plt.show()
```

```python
prod = ['A', 'B', 'C', 'D']
sale = [1000, 1500, 800, 1200]

plt.barh(prod, sale, color="orange", label="Sales 2025")
plt.xlabel("product")
plt.ylabel("Sales")
plt.legend()
plt.title("Product sales comparison")
plt.show()
```

### pie chart
plt.pie(values, labels=label_list, color=color_list, autopct=%1.2f%%)

```python
regions = ['N', 'S', 'E', 'W']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct="%1.1f%%", colors=['gold', 'skyblue', 'green', 'coral'])
plt.title("Revenue contribution by region")
plt.show()
```

### Histogram
plt.hist(data, bins=no_of_bins, color="colorname", edgecolor="black")

```python
scores = [45, 57, 67, 68, 76, 79, 82, 86, 88, 90, 93, 94, 97, 100, 105, 110]

plt.hist(scores, bins=5, color="purple", edgecolor="black")
plt.xlabel("Score range")
plt.ylabel("No of students")
plt.title("Score distribution of students")
plt.xlim(45, 110)
plt.ylim(1, 5)
plt.show()
```

### Scatter plots
plt.scatter(x, y, color="color_name", marker="marker style", label = "label_name")

```python
hr = [1,2,3,4,5,6,7,8]
score = [50,55,95,65,75,90,95, 99]

plt.scatter(hr, score, color="red", marker="^", label="Student data")
plt.xlabel("Hours studiesd")
plt.ylabel("Exam score")
plt.title("Relation between study time & exam score")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.scatter([1,2,3], [50,55,65], color="red", marker="^", label="ClassB")
plt.scatter([1,2,3], [45, 58, 62], color="green", marker="^", label="ClassA")
plt.xlabel("Hours studiesd")
plt.ylabel("Exam score")
plt.title("Comparison between two group of students")
plt.grid(True)
plt.legend()
plt.show()
```

### SUBPLOT

plt.subplot(nrows, ncols, index)
index is 1-based


subplot functions:
1. ax.set_title(), fig.suptitle()
2. plt.subplots()
3. plt.tight_layout()
4. fig.subplots_adjust()
5. plt.subplot()

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)  #1st row, 2nd col, 1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)  #1st row, 2nd col, 2nd index
plt.bar(x, y)
plt.title("Bar chart")
plt.show()
```

### fig, ax = plt.subplots(nrows, ncols, figsize(width, height))

```python
fig, ax = plt.subplots(1,2,figsize=(10, 5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x, y, color="red")
ax[0].set_title("Line Plot")

ax[1].bar(x, y, color="green")
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison of line and bar charts")

plt.tight_layout()
plt.show()
```

### Saving my visualizations
matplotlib savefig() function : saving visualizations

savefig("filename.extension", dpi=value, bbox_inches="tight)

dpi -> dots per inch(it controls image resolution)

```python
fig, ax = plt.subplots(1,2,figsize=(10, 5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x, y, color="red")
ax[0].set_title("Line Plot")

ax[1].bar(x, y, color="green")
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison of line and bar charts")

plt.tight_layout()

plt.savefig("comp.png", dpi=300, bbox_inches='tight')
plt.show()
```
